```python
import pandas as pd
movie_info = pd.read_csv('Downloads/Movie Interests DS.csv')
movie_info
```

```python
input_dataset = movie_info.drop(columns = ['Interest'])
```

```python
input_dataset
```

```python
output_dataset = movie_info['Interest']
```

```python
output_dataset
```

```python
from sklearn.tree import DecisionTreeClassifier

# Create an instance of the DecisionTreeClassifier
movie_model = DecisionTreeClassifier()

# Fit the model with the input and output datasets
movie_model.fit(input_dataset, output_dataset)

# Predict movie interest for given input data
movie_interest = movie_model.predict([[9, 1], [33, 0]])

# Display the prediction result
movie_interest
```
